```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
custom_data = pd.read_json('experiment_custom.json')
json_data = pd.read_json('experiment_json.json')
```

```python
custom_data
```

```python
custom_avg = custom_data.groupby('message').mean()
custom_avg
```

```python
json_avg = json_data.groupby('message').mean()
json_avg
```

```python
plt.plot(list(range(1, 1001,10)), custom_avg['message_size'], label='custom protocol')
plt.plot(list(range(1, 1001,10)), json_avg['message_size'], label='json protocol')
plt.legend()
plt.xlabel("number of characters in sender's message")
plt.ylabel("size of the message sent from client to server")
plt.title("size of the message sent from client to server vs number of characters in sender's message")
```

Makes sense that above would be a linear relationship

```python
plt.plot(list(range(1, 1001,10)), custom_avg['response_size'], label='custom protocol')
plt.plot(list(range(1, 1001,10)), json_avg['response_size'], label='json protocol')
plt.legend()
plt.xlabel("number of characters in sender's message")
plt.ylabel("size of the response sent from server to client")
plt.title("size of the response sent from server to client vs number of characters in sender's message")
```

```python
plt.plot(list(range(1, 1001,10)), custom_avg['time_taken'], label='custom protocol')
plt.plot(list(range(1, 1001,10)), json_avg['time_taken'], label='json protocol')
plt.legend()
plt.xlabel("number of characters in sender's message")
plt.ylabel("time taken averaged over 10 trials")
plt.title("time taken averaged over 10 trials vs number of characters in sender's message")
```
